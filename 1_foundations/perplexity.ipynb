```python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv(override=True)

# The 'perplexity' package is not available on PyPI.
# Perplexity AI's API is compatible with the OpenAI client.
# We will use the OpenAI client and set the base_url to Perplexity's API.
# Ensure PERPLEXITY_API_KEY is set in your .env file.
client = OpenAI(
    base_url="https://api.perplexity.ai",
    api_key=os.getenv("PERPLEXITY_API_KEY")
)
```

```python
# chat with grounded serach

completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "What are the major AI developments and announcements from today across the tech industry?"
        }
    ],
    model="sonar-pro"
)

print(completion.choices[0].message.content)
```
